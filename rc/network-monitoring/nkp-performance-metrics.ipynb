### Performance Metrics
***

When we think about monitoring an interface, we often are thinking about whether the interface is available or not available. 

There are many other metrics that we could gather from these interfaces that might allow us to catch a problem when it’s very small before it begins to impact the overall performance of a device.

For example, an increasing number of CRC errors may indicate a problem with a cable or with an interface, or we may be able to see the type of congestion that’s on the network and determine if the utilization may be getting too high.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Monitoring the Interface - Simple Network Management Protocol  - SNMP 
***

A lot of this we can view in the operating system of the device itself, but of course, if you’re in a very large environment, you may want to take advantage of using SNMP. 

We can query this interface over time and be able to gather metrics on these details.

We can remotely monitor every device that’s on our network and get real time access into exactly how it’s performing at any particular time.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### SNMP - Standardised Management Information Base - MIB-II Database
***

There’s a standardized management information base for SNMP called MIB-II, and almost every device supports gathering statistics from the standard MIB-II database. 

But some devices also have a proprietary MIB that allows you to gather metrics and details that are very specific to that device. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Monitoring the Interface - Link Status
***

Monitoring interface can, of course, tell you if the interface is up or down. 

If the interface is up, then everybody should be able to access that device, but if the interface is down, the problem may be with that interface or with any other component between you and that particular interface.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Monitoring the Interface - Error Rate
***

We can also monitor interfaces for errors - can can look at those error rates over time, and we can see exactly the specific error that may be occurring. 

We can see how many CRC errors may be occurring on Ethernet or WAN connection or we may be looking at detailed Ethernet metrics like runt frames or giant frames.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Monitoring the Interface - Utilisation
***

And another good monitoring statistic is to evaluate how much traffic is going through a particular interface and gather utilization details on every single interface on our network. 

Gathering this bandwidth information over time would also allow you to create some trends, so you could see exactly what direction the bandwidth may be going and if you need to plan to create additional resources for that particular service.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Monitoring the Interface - Discards / Packet Drops
***

Another set of metrics to monitor may be related to discards or packet drops. 

These errors occur when the problem isn’t associated with the packet, but instead is associated with the system’s ability to process that packet. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Monitoring the Interface - Reset
***

Might also want to monitor or log information to know if that interface is ever reset. 

This would give us an idea that that particular interface may go into an offline mode where packets are not being processed.

If we do find an interface that has been reset and has not come back into an online state, we might try to reset that interface and see if we can begin the process of communicating again to that device. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Monitoring the Interface - Interface Configuration
***

Our monitoring functions may also allow us to view the configuration of that interface.

For example, we could see the exact speed and duplex that that particular interface card is configured with. 

Could then compare that speed and duplex with the device that’s on the other end of that connection and make sure that both of those are the same. 

If the speed and duplex are not the same between the two, you’ll probably see throughput problems or late collisions occurring in your monitoring statistics.
